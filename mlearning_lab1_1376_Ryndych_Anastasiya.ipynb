### 1. Выбор исследуемого датасета. 
Для работы был выбран датасет "Средняя почасовая зарплата в США (1973-2022)" по ссылке https://www.kaggle.com/datasets/asaniczka/median-and-avg-hourly-wages-in-the-usa-1973-2022/data


### 2. Краткое описание датасета.

   Этот набор данных - информация о медиане и средней почасовой заработной плате в Соединенных Штатах в период с 1973 по 2022 год. Данные получены из библиотеки данных Института экономической политики (EPI) Штата Работающая Америка, надежного источника экономических данных.
   Для упрощения исследования было сокращено количество рассматриваемых категорий, в результате чего в датасете представлены следующие атрибуты числового типа:
* year - год
* average - средняя почасовая зарплата всего населения США
* men_average - средняя мужская оплата труда
* women_average - средняя женская оплата труда
* eur_average - средняя заработная плата людей европейского происхождения
* afr_average - средняя заработная плата людей африканского происхождения
* recent_high-school_graduate - средняя почасовая зарплата людей со средним образованием (полученным недавно)
* recent_college_graduate - средняя почасовая зарплата людей с высшим образованием (полученным недавно)

```python
import pandas as pd
import numpy as np
# Загрузка данных
data=pd.read_csv(r"C:\Users\pumpk\Desktop\university\3.1\mlearn\median_average_wages.csv")
year = ['year']
data.rename(columns = {'white_average':'eur_average', 'black_average':'afr_average'}, inplace = True )
researchColumns = ['average', 'men_average', 'women_average', 'eur_average', 'afr_average', 'recent_high-school_graduate', 'recent_college_graduate']
for element in data:
    if element not in researchColumns and element not in year:
        data=data.drop(element, axis='columns')

display(data)
```

### 3.1 Определение среднего значения и  СКО для каждого атрибута.


```python
import matplotlib.pyplot as plt
meanList = []
stdList = []
for col in researchColumns:
    meanList.append(round(data[col].mean(),2))
    stdList.append(round(data[col].std(),2))

meanStdTable = {
    'Категория рабочих': researchColumns,
    'Среднее значение': meanList,
    'СКО': stdList
}
avg_data = pd.DataFrame(meanStdTable)
print(avg_data)
```

### 3.2. Построение гистограмм распределения значений, исследование на наличие выбросов.

```python
for col in researchColumns:
    plt.figure()
    plt.title(col)
    data[col].plot(kind='hist', density=1, bins=20, alpha=.5, color='blue')
```

Построение ящиков с усами для визуализации выбросов:

```python
for col in researchColumns:
    plt.figure()
    plt.title(col)
#     sns.boxplot(data=data[col])
    
    _, bp = data[col].plot.box(return_type='both')
    outliers = [flier.get_ydata() for flier in bp["fliers"]][0]
    count = 0
    print("Выбросы в ", col)
    for elem in data[col]:  
        for i in outliers:
            if i == elem:
                print("Год: ",data['year'][count], "Значение: ", i)
        count += 1

    
```

### 3.3 Поиск пропущенных значений и их обработка в случае необходимости

```python
data.isna().sum()
```

Пропущенные значения были обнаружены в категориях "recent_high-school_graduate" и "recent_college_graduate". В качестве обработки пропущенных значений удалим данные значения из выборки.

```python
data = data[data['recent_high-school_graduate'].notna()]
data = data[data['recent_college_graduate'].notna()]
display(data)
```

### 4. Корреляция между параметрами
Вывод коэффициентов корреляции и построение матрицы графиков рассеивания для анализа корреляции

```python
import seaborn as sns

data_matrix = data
data_matrix=data_matrix.drop('year', axis='columns')
correlation_matrix = data_matrix.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# pd.plotting.scatter_matrix(data)
import seaborn as sns
sns.pairplot(data_matrix)
plt.show()
```

В следствии визуального анализа графиков рассеивания и анализа графика коэффициентов корреляции можно сделать вывод, что слабую корреляцию имеют атрибуты, связанные с образованием. В остальных случаях корреляция сильная, коэффициент близится к единице, причем характер корреляции везде положительный. 
Максимальная корреляция наблюдается у следующих наборов атрибутов: "average" и "men_average", "average" и "afr_average", "women_average" и "eur_average". Это значит, что средняя почасовая зарплата по этим параметрам росла с годами пропорциональнее остальных.
